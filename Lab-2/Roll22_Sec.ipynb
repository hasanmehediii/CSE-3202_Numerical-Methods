```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1.2 - ((x**2) / ((1 - x)**2)) * math.exp(-8.0 * x) + math.log(1 + x)
```

```python
def secant(x0, x1, tol):
    print(f"{'Iter':<5}{'x0':<15}{'x1':<15}{'xr(new)':<15}{'f(xr)':<20}{'εa (%)':<15}")
    print("-"*90)

    it = 0
    errors = []

    while True:
        it += 1
        f0, f1 = f(x0), f(x1)

        if f1 - f0 == 0:
            print("Division by zero. Not applicable")
            return None

        xr = x1 - (f1 * (x1 - x0)) / (f1 - f0)

  
        ea = abs((xr - x1) / xr) * 100
        errors.append(ea)

        print(f"{it:<5}{x0:<15.10f}{x1:<15.10f}{xr:<15.10f}{f(xr):<20.10f}"
              f"{'' if ea is None else f'{ea:.10f}'}")

        if ea is not None and ea < tol:
            break

        x0=x1
        x1=xr

    plt.figure()
    plt.plot(range(1, len(errors)+1), errors, marker='o')
    plt.xlabel('Iteration')
    plt.ylabel('Approximate  Error (%)')
    plt.title('Convergence of Approximate Error in Secant Method')
    plt.grid(True)
    plt.show()

    return xr
```

```python
def main():
    x0 = -0.3
    x1 = -0.7
    tol = 0.00001

    root = secant(x0, x1, tol)
    if root is not None:
        print(f"\nApproximate root: {root:.6f}\n")

if __name__ == "__main__":
    main()
```
