225 + 82*x - 90*x**2 + 44*x**3 - 8*x**4 + 0.7*x**5

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

Function Definition

```python
def f(x):
    return 225 + 82*x - 90*x**2 + 44*x**3 - 8*x**4 + 0.7*x**5
```

True root (high-precision reference)

```python
from scipy.optimize import brentq

if 'f' not in globals():
	def f(x):
		return 225 + 82*x - 90*x**2 + 44*x**3 - 8*x**4 + 0.7*x**5

true_root = brentq(f, -1.2, -1.0, xtol=1e-15, maxiter=1000)
print(f"True Root (reference): {true_root:.10f}\n")
```

Bisection Method

```python
xl = -1.2
xu = -1.0
es = 0.05 / 100
max_iter = 1000

assert f(xl) * f(xu) < 0, "No sign change in initial interval!"

results = []
xr_old = None

for k in range(1, max_iter + 1):
    xr = (xl + xu) / 2.0        # midpoint
    fxr = f(xr)
    fxl = f(xl)

    # True error
    et = abs((true_root - xr) / true_root) * 100

    # Approximate relative error
    if xr_old is not None:
        ea = abs((xr - xr_old) / xr) * 100
    else:
        ea = None

    results.append({
        'iter': k,
        'xl':   xl,
        'xu':   xu,
        'xr':   xr,
        'fxr':  fxr,
        'et':   et,
        'ea':   ea
    })

    ea_str = f"{ea:.10f}" if ea is not None else "     —          "
    print(f"Iter {k:3d} | xl={xl:+.10f} | xu={xu:+.10f} | "
          f"xr={xr:+.10f} | f(xr)={fxr:+.10e} | "
          f"ϵt={et:.10f}% | ϵa={ea_str}%")

    if xr_old is not None and ea < es:
        print(f"\n✅ Converged at iteration {k}: ϵa = {ea:.10f}% < {es*100}%")
        break

    if fxl * fxr < 0:
        xu = xr
    else:
        xl = xr

    xr_old = xr
```

Pretty Table

```python
print("\n" + "="*130)
header = (f"{'Iter':>4} | {'xl':>14} | {'xu':>14} | {'xr (c)':>14} | "
          f"{'f(xr)':>16} | {'ϵt (%)':>16} | {'ϵa (%)':>16}")
print(header)
print("="*130)
for r in results:
    ea_str = f"{r['ea']:16.10f}" if r['ea'] is not None else f"{'—':>16}"
    print(f"{r['iter']:>4} | {r['xl']:>14.10f} | {r['xu']:>14.10f} | "
          f"{r['xr']:>14.10f} | {r['fxr']:>16.10e} | "
          f"{r['et']:>16.10f} | {ea_str}")
print("="*130)
print(f"\nApproximate Root: {results[-1]['xr']:.10f}")
print(f"True Root:        {true_root:.10f}")
```

Plot: Absolute Error vs Iteration

```python
iters_plot = [r['iter'] for r in results if r['ea'] is not None]
ea_plot    = [r['ea']   for r in results if r['ea'] is not None]

fig, ax = plt.subplots(figsize=(10, 6))

ax.semilogy(iters_plot, ea_plot, 'o-', color='royalblue',
            linewidth=2, markersize=7, markerfacecolor='white',
            markeredgewidth=2, label='Approximate Relative Error ϵa (%)')

ax.axhline(y=es * 100, color='crimson', linestyle='--',
           linewidth=1.8, label=f'Stopping Criterion ϵs = {es*100}%')

ax.annotate(f'Converged\nIter {iters_plot[-1]}\nϵa={ea_plot[-1]:.6f}%',
            xy=(iters_plot[-1], ea_plot[-1]),
            xytext=(iters_plot[-1] - 2, ea_plot[-1] * 50),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=9, color='darkgreen',
            bbox=dict(boxstyle='round,pad=0.3', fc='lightyellow', ec='gray'))

ax.set_xlabel('Iteration Number', fontsize=13)
ax.set_ylabel('Approximate Relative Error ϵa (%)', fontsize=13)
ax.set_title('Bisection Method — Absolute Error vs Number of Iterations\n'
             r'$f(x) = 225 + 82x - 90x^2 + 44x^3 - 8x^4 + 0.7x^5$',
             fontsize=13)
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
ax.legend(fontsize=11)
ax.grid(True, which='both', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('bisection_error_plot.png', dpi=150)
plt.show()
print("\nPlot saved to bisection_error_plot.png")
```
