```python
import math
import matplotlib.pyplot as plt
```

```python
def f(m):
    return (9.81 * m / 15) * (1 - math.exp(-(15/m) * 10)) - 36
```

```python
def secant(x0, x1, tol):
    print(f"{'Iter':<5}{'x0':<15}{'x1':<15}{'xr(new)':<15}{'f(xr)':<20}{'εa (%)':<15}")
    print("-"*90)

    it = 0
    errors = []

    while True:
        it += 1
        f0, f1 = f(x0), f(x1)

        if f1 - f0 == 0:
            print("Division by zero. Not applicable")
            return None

        xr = x1 - (f1 * (x1 - x0)) / (f1 - f0)

  
        ea = abs((xr - x1) / xr) * 100
        errors.append(ea)

        print(f"{it:<5}{x0:<15.10f}{x1:<15.10f}{xr:<15.10f}{f(xr):<20.10f}"
              f"{'' if ea is None else f'{ea:.10f}'}")

        if ea is not None and ea < tol:
            break

        x0=x1
        x1=xr

    plt.figure()
    plt.plot(range(1, len(errors)+1), errors, marker='o')
    plt.xlabel('Iteration')
    plt.ylabel('Approximate  Error (%)')
    plt.title('Convergence of Approximate Error in Secant Method')
    plt.grid(True)
    plt.show()

    return xr
```

```python
def main():
    m0 = 30
    m1 = 40
    tol = 0.001

    root = secant(m0, m1, tol)
    print(f"\nApproximate root: {root:.10f}\n")

if __name__ == "__main__":
    main()
```
